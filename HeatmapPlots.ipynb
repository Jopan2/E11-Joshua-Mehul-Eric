```python
from heatmappy import Heatmapper
from PIL import Image 
from helper import getcoords
from helper import dot
from helper import getcounts
import numpy as np
```

First we want to get the coordinates relative to the edge coordinate (37.8679,-122.26555)

```python
lat_long = getcoords("gooddata/GPS_GUI_Data_2023-04-21_16:34:37_R.csv")
lats = []
for lat in lat_long["Latitude"]:
    lats.append(lat-37.8679)
longs = []
for long in lat_long["Longitude"]:
    longs.append(long-(-122.26555))
```

We now have each coordinate relative to the edge we now need them as a percentage of the total side lengths in the x and y direction 

```python
lat_side = 37.87647-37.8679
long_side = -122.25701-(-122.26555)
for i,lat in enumerate(lats):
    lats[i]=  (lat/lat_side) 
for i,long in enumerate(longs):
    longs[i] = (long/long_side) 
lats = np.array(lats)
lats = lats *1000
longs = np.array(longs)
longs = longs * 1100

```

```python
coords = list(zip(lats,longs))
coords
```

```python
cps = getcounts("gooddata/GPS_GUI_Data_2023-04-21_16:34:37_R.csv")
final_map = dot(coords,cps)
```

```python
#example_points = [(0, 0), (800,800), (200, 50), (60, 300), (170, 250)]
example_img_path = 'berkmaprrotated.png'
example_img = Image.open(example_img_path)
heatmapper = Heatmapper(point_strength=0.02,point_diameter=25)
heatmap = heatmapper.heatmap_on_img(final_map, example_img)
heatmap.show()
```
